##Exploratory Graphs

This following graph verifies that the coverage / score increases over time. It is confusing that the low entry values also keep popping up. It seems that modkit is not able to account for insertions / deletions in the pileup script? This definietely needs further investigation

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the path to your BED file
bed_file_path = 'HCV_TotalRNA_004_ALIGNED_SIGNALS.BASECALLED_mod.bed'

# Define the column names for the BED file
bed_columns = [
    'chrom', 'start', 'end', 'modified_base_code', 'score', 'strand',
    'start_compatibility', 'end_compatibility', 'color', 'Nvalid_cov',
    'fraction_modified', 'Nmod', 'Ncanonical', 'Nother_mod', 'Ndelete',
    'Nfail', 'Ndiff', 'Nnocall'
]

# Read the BED file
bed_df = pd.read_csv(bed_file_path, sep='\t', names=bed_columns)

# Plotting start position against score
plt.figure(figsize=(10, 6))
plt.scatter(bed_df['start'], bed_df['score'], alpha=0.5, edgecolors='w', linewidth=0.5)
plt.title('Start Position vs Score')
plt.xlabel('Start Position')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

graphed all of the fractions of either m6a (a) or pseudouridine (17802) - more diversity once there is more coverage since there is a much larger population of reads to sample for each position.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Define the path to your BED file
bed_file_path = 'HCV_TotalRNA_004_ALIGNED_SIGNALS.BASECALLED_mod.bed'

# Define the column names for the BED file
bed_columns = [
    'chrom', 'start', 'end', 'modified_base_code', 'score', 'strand',
    'start_compatibility', 'end_compatibility', 'color', 'Nvalid_cov',
    'fraction_modified', 'Nmod', 'Ncanonical', 'Nother_mod', 'Ndelete',
    'Nfail', 'Ndiff', 'Nnocall'
]

# Read the BED file
bed_df = pd.read_csv(bed_file_path, sep='\t', names=bed_columns)

# Select the entry with the highest score for each start position
bed_df_highest_score = bed_df.loc[bed_df.groupby('start')['score'].idxmax()]

# Convert the modified_base_code to categorical codes
bed_df_highest_score['mod_base_code_cat'] = bed_df_highest_score['modified_base_code'].astype('category').cat.codes

# Create a scatter plot of fraction_modified across positions, colored by modified_base_code
plt.figure(figsize=(12, 8))
scatter = plt.scatter(
    bed_df_highest_score['start'],
    bed_df_highest_score['fraction_modified'],
    c=bed_df_highest_score['mod_base_code_cat'],
    cmap='viridis',
    alpha=0.7,
    edgecolors='w',
    linewidth=0.5
)

# Create a legend
unique_codes = bed_df_highest_score['modified_base_code'].astype('category').cat.codes.unique()
unique_labels = bed_df_highest_score['modified_base_code'].astype('category').cat.categories

# Add legend for modified base codes
legend_elements = [plt.Line2D([0], [0], marker='o', color='w', label=label,
                              markerfacecolor=plt.cm.viridis(code / max(unique_codes)), markersize=10)
                   for code, label in zip(unique_codes, unique_labels)]
plt.legend(handles=legend_elements, title='Modified Base Code')

plt.title('Fraction Modified across Positions')
plt.xlabel('Start Position')
plt.ylabel('Fraction Modified')
plt.grid(True)
plt.show()
```

This looks like a more realistic profile.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Define the path to your BED file
bed_file_path = 'HCV_TotalRNA_004_ALIGNED_SIGNALS.BASECALLED_mod.bed'

# Define the column names for the BED file
bed_columns = [
    'chrom', 'start', 'end', 'modified_base_code', 'score', 'strand',
    'start_compatibility', 'end_compatibility', 'color', 'Nvalid_cov',
    'fraction_modified', 'Nmod', 'Ncanonical', 'Nother_mod', 'Ndelete',
    'Nfail', 'Ndiff', 'Nnocall'
]

# Read the BED file
bed_df = pd.read_csv(bed_file_path, sep='\t', names=bed_columns)

# Filter for start positions onward from 9,000 and score of at least 10
bed_df_filtered = bed_df[(bed_df['start'] >= 9000) & (bed_df['score'] >= 10)]

# Select the entry with the highest score for each start position
bed_df_highest_score = bed_df_filtered.loc[bed_df_filtered.groupby('start')['score'].idxmax()]

# Convert the modified_base_code to categorical codes
bed_df_highest_score['mod_base_code_cat'] = bed_df_highest_score['modified_base_code'].astype('category').cat.codes

# Create a scatter plot of fraction_modified across positions, colored by modified_base_code
plt.figure(figsize=(12, 8))
scatter = plt.scatter(
    bed_df_highest_score['start'],
    bed_df_highest_score['fraction_modified'],
    c=bed_df_highest_score['mod_base_code_cat'],
    cmap='viridis',
    alpha=0.7,
    edgecolors='w',
    linewidth=0.5
)

# Create a legend
unique_codes = bed_df_highest_score['modified_base_code'].astype('category').cat.codes.unique()
unique_labels = bed_df_highest_score['modified_base_code'].astype('category').cat.categories

# Add legend for modified base codes
legend_elements = [plt.Line2D([0], [0], marker='o', color='w', label=label,
                              markerfacecolor=plt.cm.viridis(code / max(unique_codes)), markersize=10)
                   for code, label in zip(unique_codes, unique_labels)]
plt.legend(handles=legend_elements, title='Modified Base Code')

plt.title('Fraction Modified across Positions')
plt.xlabel('Start Position')
plt.ylabel('Fraction Modified')
plt.grid(True)
plt.show()
```
