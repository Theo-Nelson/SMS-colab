##Unified Processing Script - Dorado New M6A All-Context HAC Model (HCV)

```python
%%bash

# Convert blow5 to fast5 for both samples
/content/slow5tools-v1.1.0/slow5tools s2f /content/HCV_TotalRNA_004_ALIGNED_SIGNALS.blow5 -d /content/HCV_TotalRNA_004_ALIGNED_SIGNALS.fast5
/content/slow5tools-v1.1.0/slow5tools s2f /content/HCV_IVT_004_ALIGNED_SIGNALS.blow5 -d /content/HCV_IVT_004_ALIGNED_SIGNALS.fast5

# Basecalling for both samples
cd /content/dorado-0.7.0-linux-x64/bin
./dorado basecaller rna004_130bps_hac@v5.0.0 /content/HCV_TotalRNA_004_ALIGNED_SIGNALS.fast5/HCV_TotalRNA_004_ALIGNED_SIGNALS.fast5 --modified-bases m6A > /content/HCV_TotalRNA_004_ALIGNED_SIGNALS_MOD.BASECALLED.bam
./dorado basecaller rna004_130bps_hac@v5.0.0 /content/HCV_IVT_004_ALIGNED_SIGNALS.fast5/HCV_IVT_004_ALIGNED_SIGNALS.fast5 --modified-bases m6A > /content/HCV_IVT_004_ALIGNED_SIGNALS_MOD.BASECALLED.bam

# Convert BAM to FASTQ for both samples
/content/samtools-1.20/samtools fastq -T "*" /content/HCV_TotalRNA_004_ALIGNED_SIGNALS_MOD.BASECALLED.bam > /content/HCV_TotalRNA_004_ALIGNED_SIGNALS_MOD.BASECALLED.fastq
/content/samtools-1.20/samtools fastq -T "*" /content/HCV_IVT_004_ALIGNED_SIGNALS_MOD.BASECALLED.bam > /content/HCV_IVT_004_ALIGNED_SIGNALS_MOD.BASECALLED.fastq

# Align the reads for both samples
minimap2 -ax splice -t 16 -uf -k14 -y /content/hcv.fasta /content/HCV_TotalRNA_004_ALIGNED_SIGNALS_MOD.BASECALLED.fastq > /content/HCV_TotalRNA_004_ALIGNED_SIGNALS.ALIGNED_MOD.sam
minimap2 -ax splice -t 16 -uf -k14 -y /content/hcv.fasta /content/HCV_IVT_004_ALIGNED_SIGNALS_MOD.BASECALLED.fastq > /content/HCV_IVT_004_ALIGNED_SIGNALS.ALIGNED_MOD.sam

# Convert SAM to sorted BAM and index for both samples
/content/samtools-1.20/samtools view -S -b /content/HCV_TotalRNA_004_ALIGNED_SIGNALS.ALIGNED_MOD.sam > /content/HCV_TotalRNA_004_ALIGNED_SIGNALS.ALIGNED_MOD.bam
/content/samtools-1.20/samtools view -S -b /content/HCV_IVT_004_ALIGNED_SIGNALS.ALIGNED_MOD.sam > /content/HCV_IVT_004_ALIGNED_SIGNALS.ALIGNED_MOD.bam

/content/samtools-1.20/samtools sort /content/HCV_TotalRNA_004_ALIGNED_SIGNALS.ALIGNED_MOD.bam -o /content/HCV_TotalRNA_004_ALIGNED_SIGNALS.ALIGNED_MOD.sorted.bam
/content/samtools-1.20/samtools sort /content/HCV_IVT_004_ALIGNED_SIGNALS.ALIGNED_MOD.bam -o /content/HCV_IVT_004_ALIGNED_SIGNALS.ALIGNED_MOD.sorted.bam

/content/samtools-1.20/samtools index /content/HCV_TotalRNA_004_ALIGNED_SIGNALS.ALIGNED_MOD.sorted.bam
/content/samtools-1.20/samtools index /content/HCV_IVT_004_ALIGNED_SIGNALS.ALIGNED_MOD.sorted.bam

# Generate BED files for both samples
cd /content/dist/
./modkit pileup /content/HCV_TotalRNA_004_ALIGNED_SIGNALS.ALIGNED_MOD.sorted.bam /content/HCV_TotalRNA_004_ALIGNED_SIGNALS.BASECALLED_mod.bed --ref /content/hcv.fasta
./modkit pileup /content/HCV_IVT_004_ALIGNED_SIGNALS.ALIGNED_MOD.sorted.bam /content/HCV_IVT_004_ALIGNED_SIGNALS.BASECALLED_mod.bed --ref /content/hcv.fasta
```

##Analysis - Dorado New M6A All-Context HAC Model (HCV)

There seems to high correspondence between the values in the total HCV and IVT HCV for the HAC m6a all-context model.

There is only one potential m6a site identified with a reasonable proportion of difference [68.97 Total vs. 33.71 IVT]

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the paths to your BED files
total_rna_bed_path = 'HCV_TotalRNA_004_ALIGNED_SIGNALS.BASECALLED_mod.bed'
ivt_bed_path = 'HCV_IVT_004_ALIGNED_SIGNALS.BASECALLED_mod.bed'

# Define the column names for the BED files
bed_columns = [
    'chrom', 'start', 'end', 'modified_base_code', 'score', 'strand',
    'start_compatibility', 'end_compatibility', 'color', 'Nvalid_cov',
    'fraction_modified', 'Nmod', 'Ncanonical', 'Nother_mod', 'Ndelete',
    'Nfail', 'Ndiff', 'Nnocall', 'sample'
]

# Read the BED files
total_rna_df = pd.read_csv(total_rna_bed_path, sep='\t', names=bed_columns)
ivt_df = pd.read_csv(ivt_bed_path, sep='\t', names=bed_columns)

# Add sample column
total_rna_df['sample'] = 'TotalRNA'
ivt_df['sample'] = 'IVT'

# Filter for score of at least 10
total_rna_df = total_rna_df[total_rna_df['score'] >= 10]
ivt_df = ivt_df[ivt_df['score'] >= 10]

# Filter for m6A_DRACH
total_rna_df = total_rna_df[total_rna_df['modified_base_code'] == 'a']
ivt_df = ivt_df[ivt_df['modified_base_code'] == 'a']

# Merge dataframes on 'chrom' and 'start' to ensure they have shared points
merged_df = pd.merge(
    total_rna_df[['chrom', 'start', 'fraction_modified']].rename(columns={'fraction_modified': 'TotalRNA_fraction_modified'}),
    ivt_df[['chrom', 'start', 'fraction_modified']].rename(columns={'fraction_modified': 'IVT_fraction_modified'}),
    on=['chrom', 'start']
)

# Function to create scatter plot
def plot_modifications(df, title, filename):
    # Create a scatter plot of TotalRNA_fraction_modified (x-axis) vs. IVT_fraction_modified (y-axis)
    plt.figure(figsize=(12, 8))
    plt.scatter(
        df['TotalRNA_fraction_modified'],
        df['IVT_fraction_modified'],
        alpha=0.7,
        edgecolors='w',
        linewidth=0.5,
        color='purple'
    )

    plt.title(title)
    plt.xlabel('TotalRNA Fraction Modified')
    plt.ylabel('IVT Fraction Modified')
    plt.grid(True)
    plt.savefig(filename)
    plt.show()

# Plot m6A_DRACH
plot_modifications(merged_df, 'm6A Fraction Modified: TotalRNA vs IVT', 'm6a_scatter_plot_v3.png')

# Get top 10 positions where IVT is higher than TotalRNA
ivt_higher = merged_df[merged_df['IVT_fraction_modified'] > merged_df['TotalRNA_fraction_modified']]
top_10_ivt_higher = ivt_higher.nlargest(10, 'IVT_fraction_modified')

# Get top 10 positions where TotalRNA is higher than IVT
total_higher = merged_df[merged_df['TotalRNA_fraction_modified'] > merged_df['IVT_fraction_modified']]
top_10_total_higher = total_higher.nlargest(10, 'TotalRNA_fraction_modified')

# Print the results
print("Top 10 positions where IVT is higher than TotalRNA:")
print(top_10_ivt_higher[['chrom', 'start', 'IVT_fraction_modified', 'TotalRNA_fraction_modified']])

print("\nTop 10 positions where TotalRNA is higher than IVT:")
print(top_10_total_higher[['chrom', 'start', 'TotalRNA_fraction_modified', 'IVT_fraction_modified']])
```
