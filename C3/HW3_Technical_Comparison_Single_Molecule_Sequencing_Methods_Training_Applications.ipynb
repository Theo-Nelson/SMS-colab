The following script is a modification on the picoAmps script which performs combined sub-sampling and conversion.

```python
%%bash

input_blow5="/content/HCV_IVT_002_REP1_ALIGNED_SIGNALS.blow5"
output_folder="/content/subsamples"
mkdir -p $output_folder

# Loop for three sub-samplings
for i in {1..3}
do
    # Sub-sample 500 random reads
    /content/slow5tools-v1.1.0/slow5tools skim --rid $input_blow5 | sort -R | head -20 > $output_folder/rand_500_rid_$i.txt
    /content/slow5tools-v1.1.0/slow5tools get $input_blow5 --list $output_folder/rand_500_rid_$i.txt -o $output_folder/HCV_IVT_002_REP1_ALIGNED_SIGNALS_SAMPLE_$i.blow5

    # Convert BLOW5 to SLOW5
    /content/slow5tools-v1.1.0/slow5tools view $output_folder/HCV_IVT_002_REP1_ALIGNED_SIGNALS_SAMPLE_$i.blow5 -o $output_folder/HCV_IVT_002_REP1_ALIGNED_SIGNALS_SAMPLE_$i.slow5

    # Conversion of signals to picoAmps using awk
    awk 'BEGIN {FS="\t"; OFS="\t"}
        !/^[@#]/ {
            read_id = $1
            digitisation = $3
            offset = $4
            range = $5
            raw_signal = $8
            n = split(raw_signal, signal_array, ",")
            printf "%s\t", read_id
            for (i=1; i<=n; i++) {
                signal_in_pico_ampere = (signal_array[i] + offset) * range / digitisation
                printf "%f,", signal_in_pico_ampere
            }
            printf "\n"
        }' $output_folder/HCV_IVT_002_REP1_ALIGNED_SIGNALS_SAMPLE_$i.slow5 > $output_folder/converted_$i.tsv
done

# Combine all samples for PCA analysis
cat $output_folder/converted_*.tsv > $output_folder/all_converted_signals.tsv
```

We notice that the signals do not cluster amongst themselves, suggesting that the random sub-samplings are giving us random samples from within the same population.

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('/content/subsamples/all_converted_signals.tsv', sep='\t', header=None)

# Simple preprocessing: extract numerical data, handle missing values
signals = data.iloc[:, 1].str.split(',', expand=True)
signals = signals.apply(pd.to_numeric, errors='coerce')
signals.fillna(method='ffill', axis=1, inplace=True)

# Perform PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(signals)

# Plot the results
plt.figure(figsize=(8, 6))
colors = ['r', 'g', 'b']  # Color for each subset
for i in range(3):
    start_index = i * 20  # Correct start index for each subset
    end_index = start_index + 20  # Correct end index for each subset
    plt.scatter(principal_components[start_index:end_index, 0], principal_components[start_index:end_index, 1], c=colors[i], label=f'Subset {i+1}')
plt.title('PCA of Nanopore Signals')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

The signals are not identical, since they are all different lengths [see the length distributions below]. Ideally, the nanopore adapters ligate to the 3' end, meaning that all of the signals have the same start site within the HCV molecules.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('/content/subsamples/all_converted_signals.tsv', sep='\t', header=None)

# Initialize a list to store signal lengths for each subset
signal_lengths = []
min_lengths = []

# Extract signal lengths and find minimums
for i in range(3):
    start_index = i * 20  # Start index for each subset
    end_index = start_index + 20  # End index for each subset
    subset = data.iloc[start_index:end_index, 1].str.split(',', expand=True)
    lengths = subset.apply(lambda x: x.notnull().sum(), axis=1)
    signal_lengths.append(lengths)
    min_lengths.append(lengths.min())

# Print the minimum lengths
labels = ['HCV IVT 002 REP1', 'HCV IVT 002 REP2', 'HCV IVT 004']
for label, min_length in zip(labels, min_lengths):
    print(f'Minimum length in {label}: {min_length}')

# Plot the signal length distributions
plt.figure(figsize=(10, 6))
bins = np.linspace(0, max(max(sl) for sl in signal_lengths), 50)  # Create bins up to the maximum length found
for lengths, label in zip(signal_lengths, labels):
    plt.hist(lengths, bins=bins, alpha=0.5, label=label)
plt.xlabel('Signal Length')
plt.ylabel('Frequency')
plt.title('Signal Length Distribution by Sample')
plt.legend()
plt.show()
```

Nanopore sequencing then proceeds 3' to 5' through the pore. Therefore, we examine the sequence lengths and re-perform the above analysis, considering only the 'shared' nucleotide length at the beginning of the signal, in this case 10,897 signal entries.

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('/content/subsamples/all_converted_signals.tsv', sep='\t', header=None)

# Trim each signal to the first 10897 entries
data[1] = data[1].apply(lambda x: ','.join(x.split(',')[:10897]))

# Simple preprocessing: extract numerical data, handle missing values
signals = data[1].str.split(',', expand=True)
signals = signals.apply(pd.to_numeric, errors='coerce')
signals.fillna(method='ffill', axis=1, inplace=True)

# Perform PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(signals)

# Plot the results
plt.figure(figsize=(8, 6))
colors = ['r', 'g', 'b']  # Color for each subset
labels = ['HCV IVT 002 REP1', 'HCV IVT 002 REP2', 'HCV IVT 004']  # Labels based on the sample origin
num_entries = len(data) // 3  # Assuming equal division of entries among three samples
for i in range(3):
    start_index = i * num_entries
    end_index = start_index + num_entries
    plt.scatter(principal_components[start_index:end_index, 0], principal_components[start_index:end_index, 1], c=colors[i], label=labels[i])
plt.title('PCA of Nanopore Signals')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title="Sample Source")
plt.show()
```

```python
import pandas as pd
import numpy as np
from numpy.fft import fft
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('/content/subsamples/all_converted_signals.tsv', sep='\t', header=None)

# Assume data is evenly divided among three samples, add source labels
sample_labels = ['HCV IVT 002 REP1', 'HCV IVT 002 REP2', 'HCV IVT 004']
num_samples = len(data) // 3
data['Source'] = [sample_labels[i // num_samples] for i in range(len(data))]

# Trim each signal to the first 10897 entries, convert to numeric, and handle non-numeric data
def process_signal(x):
    try:
        return np.array([float(i) for i in x.split(',')[:10897] if i.strip() != ''], dtype=float)
    except ValueError:
        return np.array([np.nan]*10897)  # Return NaNs if conversion fails

data[1] = data[1].apply(process_signal)

# Perform DFT on each signal and store only the real part of the first half (symmetry in real signals)
data['DFT'] = data[1].apply(lambda x: np.abs(fft(x)[:len(x)//2]))

# Convert the list of DFT arrays into a DataFrame, maintaining the source information
dft_features = pd.DataFrame(data['DFT'].tolist())
dft_features['Source'] = data['Source']

# Drop any rows with NaNs
dft_features.dropna(inplace=True)

# Calculate the mean DFT value for each frequency bin for each source
mean_dft_by_source = dft_features.groupby('Source').mean()

# Plotting the average DFT values on a logarithmic scale
plt.figure(figsize=(14, 8))
# Generate frequency bin indices
frequency_bins = np.arange(mean_dft_by_source.shape[1])
for source, color in zip(mean_dft_by_source.index, ['r', 'g', 'b']):
    plt.plot(frequency_bins, mean_dft_by_source.loc[source], label=source, color=color)

plt.title('Average DFT Frequencies by Sample Source on Log Scale')
plt.xlabel('Frequency Bin')
plt.ylabel('Average DFT Magnitude')
plt.yscale('log')  # Set the y-axis to logarithmic scale
plt.legend(title="Sample Source")
plt.grid(True, which="both", ls="--")  # Enhanced grid for log scale
plt.show()
```
