## Problem 1

My way of thinking about sequencing quality:

Oxford Nanopore reports that it's flow cells perform signal acquistion at 2-20 kHz and that the pore processes 400-450 bases per second. This means that we should have between 4-50 measurements for every base. Thus, the
difference between pairs of conseeuctive signals should be, on average, zero.
If not, then we have more noise interfering with the signal and thus signal quality.

```python
%%bash


# Define the input SLOW5 file
input_slow5="/content/HCV_IVT_004_ALIGNED_SIGNALS.blow5"
subsample_input_slow5="HCV_IVT_004_ALIGNED_SIGNALS_500_RANDOM_READS.blow5"

for n in 1 2 3;
do

  # Output file to store the converted signals
  output_file="HCV_IVT_004_aligned_converted_signals_"${n}".txt"

  # Check if the output file exists and remove it to start fresh
  [ -f "$output_file" ] && rm "$output_file"

  slow5tools view $input_slow5 | grep -v '^[#@]' | shuf -n 500 > $subsample_input_slow5

  # Read each line of the SLOW5 file, skip header lines starting with # or @
  awk 'BEGIN {FS="\t"}
      !/^[@#]/ {
          # Extract fields based on their positions:
          # Assuming read_id is $1, digitisation is $3, offset is $4, range is $5, and raw_signal is $8
          read_id = $1
          digitisation = $3
          offset = $4
          range = $5
          raw_signal = $8

          # Split the raw_signal field into individual samples
          n = split(raw_signal, signal_array, ",")

          # Convert each raw signal to picoamperes and store in a new array
          for (i=1; i<=n; i++) {
              signal_in_pico_ampere[i] = (signal_array[i] + offset) * range / digitisation
          }

          # Print the read_id and the converted signals; adjust output as needed
          printf "%s\t", read_id
          #for (i=1; i<=n; i++) {
              #printf "%f,", signal_in_pico_ampere[i]
          #}
          signal_in_pico_ampere_sum_of_differences=0
          signal_in_pico_ampere_sum=signal_in_pico_ampere[1]
          for (i=2; i<=n; i++) {
              signal_in_pico_ampere_sum_of_differences += signal_in_pico_ampere[i]-signal_in_pico_ampere[i-1]
              signal_in_pico_ampere_sum += signal_in_pico_ampere[i]
          }
          printf "%f", signal_in_pico_ampere_sum_of_differences/n
          print "\t", signal_in_pico_ampere_sum/n
          printf "\n"
      }' $subsample_input_slow5 > $output_file

done

```

```python
%%R
subsample_1 <- read.csv("HCV_IVT_004_aligned_converted_signals_1.txt", sep='\t', header=FALSE)
subsample_2 <- read.csv("HCV_IVT_004_aligned_converted_signals_2.txt", sep='\t', header=FALSE)
subsample_3 <- read.csv("HCV_IVT_004_aligned_converted_signals_3.txt", sep='\t', header=FALSE)

subsample_1$group <- 1
subsample_2$group <- 2
subsample_3$group <- 3

concat_frame <- rbind(subsample_1,subsample_2)
concat_frame <- rbind(concat_frame,subsample_3)
colnames(concat_frame) <- c("read_id", "average_signal_diff","average_signal", "group")
concat_frame$group <- as.factor(concat_frame$group)
concat_frame$sample <- "HCV_IVT_004"
tail(concat_frame)
```

```python
%%R
my_comparisons <- list( c("1", "2"), c("2", "3"), c("1", "3") )
p <- ggplot(concat_frame, aes(x=group, y=average_signal_diff)) +
  geom_boxplot() + stat_compare_means(comparisons = my_comparisons, method = "t.test")
p
```

```python
%%R
my_comparisons <- list( c("1", "2"), c("2", "3"), c("1", "3") )
p <- ggplot(concat_frame, aes(x=group, y=average_signal)) +
  geom_boxplot() + stat_compare_means(comparisons = my_comparisons, method = "t.test")
p
```

The random subsamples of 500 reads each show similar (at the very least we have no significant evidence that they are different) distributions of both the average signal and the signal-to-signal difference. Thus, it seems that subsampling is a viable strategy for exploratory analysis.

The boxplots of the average signals and differences between consecutive signal pairs across subsamples shows subsamples from the same sample vary less than subsamples from different samples. Additionally, the mean of the average signals differ between technical replicates, indicating a batch effect. Furthermore, the standard deviation of average signals differ between samples, perhaps indicating differences in the heterogeneity of sequences captured.
